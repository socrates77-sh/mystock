```python
%matplotlib inline
from all_shares import *
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('result_all.csv', index_col=0, encoding='gb2312')
```

### 基本信息

```python
print('统计总数:\t\t%s' % df.index.size)
s = df['bs-250d'].value_counts()[0]
b = df['bs-250d'].value_counts()[1]
print('B总数:\t\t\t%d\t(%0.2f%%)' % (b, 100*b/(s+b)))
print('S总数:\t\t\t%d\t(%0.2f%%)' % (s, 100*s/(s+b)))
print('250d平均收益率均值:\t%0.2f' % df['avg-250d'].mean())
print('250d最大收益率均值:\t%0.2f' % df['max-250d'].mean())
print('500d平均收益率均值:\t%0.2f' % df['avg-500d'].mean())
print('500d最大收益率均值:\t%0.2f' % df['max-500d'].mean())
```

### BS分布

```python
df['bs-250d'].value_counts().plot(kind='barh', figsize=(10,1))
```

### 平均/最大收益分布

```python
fig, ax = plt.subplots(1, 2)
df[['avg-250d', 'max-250d']].plot(kind='kde', ax=ax[0], figsize=(15,3), grid=True, title='250d')
df[['avg-500d', 'max-500d']].plot(kind='kde', ax=ax[1], figsize=(15,3), grid=True, title='500d')
```

### 最佳策略分布

```python
s1 = df['strategy-250d'].value_counts()
s2 = df['strategy-500d'].value_counts()
df_count = pd.DataFrame([s1, s2])
df_count.transpose().plot(kind='barh', figsize=(10,5))
```
